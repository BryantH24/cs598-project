Fetch and Version Postsecondary School Locations Data

```python
import requests
from pathlib import Path
from datetime import datetime
```

```python
data_dir = Path("../../postsecondary_school_locations")
data_dir.mkdir(exist_ok=True)
```

Download and Version Data

```python

dataset_url = "https://data-nces.opendata.arcgis.com/api/download/v1/items/296839772bf14df29c290202f8547ff1/csv?layers=0"
dataset_name = "postsecondary_school_locations_2020-21"

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"{dataset_name}_{timestamp}.csv"
file_path = data_dir / filename


response = requests.get(dataset_url, timeout=300)
response.raise_for_status()

with open(file_path, 'wb') as f:
    f.write(response.content)

print(file_path)
```
