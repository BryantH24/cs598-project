Fetch and Version Football Data

```python
import subprocess
from pathlib import Path
from datetime import datetime

data_dir = Path("../../college_football_data")
data_dir.mkdir(exist_ok=True) 
```

Download College Football Data

```python
dataset_slug = "cviaxmiwnptr/college-football-team-stats-2002-to-january-2024"
download_path = str(data_dir)

result = subprocess.run([
    "kaggle", "datasets", "download",
    "-d", dataset_slug,
    "-p", download_path,
    "--unzip"
], check=True, capture_output=True, text=True)
```

Version the CSV

```python
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
original_csv = data_dir / "cfb_box-scores_2002-2024.csv"
versioned_csv = data_dir / f"cfb_box-scores_2002-2024_{timestamp}.csv"

if original_csv.exists():
    original_csv.rename(versioned_csv)
```

Limit Data If Specified

```python
import pandas as pd
import os

data_subset_rows = os.environ.get('DATA_SUBSET_ROWS')
if data_subset_rows is not None:
    print("Limiting data to", data_subset_rows, "rows")
    data_subset_rows = int(data_subset_rows)
    df = pd.read_csv(versioned_csv)
    df_limited = df.head(data_subset_rows)
    df_limited.to_csv(versioned_csv, index=False)
```
